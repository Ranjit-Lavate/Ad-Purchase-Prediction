```python
# we have to identifed on the basis of given information if the customer is going to purchase the ad or not
df["Purchased"].value_counts()
```

```python
#splitting the input and the target column
x=df.iloc[:,1:4]
y=df.iloc[:,4:]
```

```python
x
```

```python
y
```

```python
#splitting the data into train and test 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=56)

print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
x_train
```

```python
#Encoding gender column
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(drop="first",sparse_output=False)

one_hot_encoded=encoder.fit_transform(x_train[["Gender"]])
one_hot_encoded
x_train_trans=pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(["Gender"]),index=x_train["Gender"].index)
x_train_trans
```

```python

one_hot_encoded=encoder.fit_transform(x_test[["Gender"]])
one_hot_encoded
x_test_trans=pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(["Gender"]),index=x_test["Gender"].index)
x_test_trans
```

```python
# merging the encoded feature into train and test data.

x_train["Gender"]=x_train_trans
x_test["Gender"]=x_test_trans
```

```python
x_train
```

```python
x_test
```

```python
# deciding the value of k

# sqaure root of count of training data points
import numpy as np
np.sqrt(x_train.shape[0])
```

```python
#model building
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=17)
knn.fit(x_train,y_train)
```

```python
# prediction:

y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# finding the accuracy of the model

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
